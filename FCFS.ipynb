```python
def FCFS(processes: dict):
    processes_names = processes.keys()
    process_queue = []
    executing_process = None
    for quantum in range(10000):
        if executing_process != None:
            if processes[executing_process][0] + processes[executing_process][3] + processes[executing_process][1] == quantum:
                processes[executing_process][4] = (processes[executing_process][0] + processes[executing_process][3] + processes[executing_process][1])
                executing_process = process_queue[0]
                process_queue.pop(0)
        for process in processes_names:
            print(process_queue, quantum)
            if processes[process][0] == quantum:
                process_queue.append(process)
            for i in process_queue:
                processes[i][3] += 1
        if executing_process != None and len(process_queue) != 0:
            executing_process = process_queue[0]
            process_queue.pop(0)
        else:
            continue
                
```
